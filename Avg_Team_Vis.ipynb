### Visualizing team data

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn  as sb
```

```python
df = pd.read_csv("Team_batchUdacity_TOTAL.csv")
df.drop("Unnamed: 0", axis=1, inplace=True)
attributes = df.columns.values.tolist()
#df.drop("Unnamed: 0.1", axis=1, inplace=True)
df.drop("MatchID", axis=1, inplace=True)
#df.drop("TeamAvgKillDist", axis=1, inplace=True)
print(attributes)
```

```python
def replaceStr(string):
    
    if string == "Alpha":
        return 0
    else:
        return 1
    
df["Unnamed: 0.1"] = df["Unnamed: 0.1"].apply(replaceStr)
```

```python
print(df.head(n=6))
```

```python
sb.set()
colors = ["greyish", "true green", "ocean blue"]
sb.pairplot(df, hue="standing", palette=sb.xkcd_palette(colors), 
            vars=["TeamAvgScorePerLife", "TeamAvgKD", "Unnamed: 0.1", "TeamAvgCR"])
```

```python
#colors = ["greyish", "true green", "ocean blue"]

g = sb.PairGrid(df, hue='standing', palette=sb.xkcd_palette(colors), 
                vars=["TeamAvgHeavyKills", "TeamAvgOtherKills", "TeamAvgPrimaryKills", "TeamAvgSecondaryKills", "TeamAvgCR"])
g = g.map_diag(plt.hist, histtype="step", linewidth=2)
g = g.map_offdiag(plt.scatter)
g = g.add_legend()
g
```
